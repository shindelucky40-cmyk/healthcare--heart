```python
for c in df.columns:
    print(c, df[c].nunique(), sorted(df[c].unique()[:10]))

# Example encoding:
cat_cols = ['sex','cp','fbs','restecg','exang','slope','ca','thal']   # adapt to your dataset's column names
for c in cat_cols:
    if c in df.columns:
        df[c] = df[c].astype('category')
        
# One-hot encode multi-category columns (drop first to avoid multicollinearity)
df = pd.get_dummies(df, columns=[c for c in cat_cols if df[c].nunique()>2], drop_first=True)
```

```python
df.head(222)
```

```python
from sklearn.preprocessing import StandardScaler

num_cols = ["age", "trestbps", "chol", "thalach", "oldpeak"]

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
df = df.astype(int)
```
